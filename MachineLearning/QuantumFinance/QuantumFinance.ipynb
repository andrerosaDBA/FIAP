Foram encontrados registros com dados com caracteres especiais, nulos, vázio, etc. Para os dados nessas condições, foram feitos tratamentos levando em consideração, o "Customer_ID", com base nele, foram feitas as correções.

```python
def preencher_valores_nulo(df_credito):
    #Ordenar o DataFrame por Customer_ID
    df_credito.sort_values(by = ['Customer_ID'])

    #Preencher valores vázios com o primeiro valor não nulo
    df_credito['Monthly_Inhand_Salary'] = df_credito.groupby(['Customer_ID'], group_keys=True )[['Monthly_Inhand_Salary']].transform(
        lambda x: x.ffill().bfill()
    )
    
    df_credito['Name'] = df_credito.groupby(['Customer_ID'], group_keys=True )[['Name']].transform(
        lambda x: x.ffill().bfill()
    )

    #Primeiro alteramos os valores que possuem '_____' na coluna 'Occupation' para Nan, depois substití para o 1° valor válido por cliente.
    df_credito['Occupation'] = df_credito['Occupation'].replace('_______', np.nan)
    df_credito['Occupation'] = df_credito.groupby(['Customer_ID'], group_keys=True )[['Occupation']].transform(
        lambda x: x.ffill().bfill()
    )

    #Primeiro alteramos os valores que possuem '#F%$D@*&8' na coluna 'SSN' para Nan, depois substití para o 1° valor válido por cliente.
    df_credito['SSN'] = df_credito['SSN'].replace('#F%$D@*&8', np.nan)
    df_credito['SSN'] = df_credito.groupby(['Customer_ID'], group_keys=True )[['SSN']].transform(
        lambda x: x.ffill().bfill()
    )

    return df_credito

df_preenchido = preencher_valores_nulo(df_credito)

print("\nDataFrame após preenchimento dos valores vazios:")
print(df_credito)
```
