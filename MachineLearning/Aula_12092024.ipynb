```python
import pandas as pd
import seaborn as sns

treino = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/data/main/salario_conta%20.csv')
treino.head(3)
```

```python
sns.scatterplot(x=treino['salario'],y=treino['saldo'])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5 )
kmeans.fit(treino)
```

```python
sns.scatterplot(x=treino['salario'],y=treino['saldo'],hue=kmeans.labels_,palette='rainbow')
```

```python
(kmeans.score(treino)*-1)**0.5
```

```python
teste = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/data/main/salario_conta_ts.csv')
teste.head(2)
```

```python
prevs=kmeans.predict(teste)
prevs
```

```python
sns.scatterplot(x=teste['salario'],y=teste['saldo'],hue=prevs,palette='rainbow')
```

```python
(kmeans.score(teste)*-1)**0.5
```
